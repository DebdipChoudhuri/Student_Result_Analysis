```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.isnull().sum()
```

# DROP UNAMED COLUMN

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

# GENDER DISTRIBUTION

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender", hue = "Gender")
plt.title("Gender Distribution")

for i in ax.containers:
    ax.bar_label(i,)

plt.show()
```

# From the above chart we have analysed that:
# the number of females in the data is more than the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb, annot = True)
plt.title(" Relationship between Parent's Education and Student's Score")
plt.show()
```

# from the above chart we have concluded that the education of the parents have a good impact on their child's scores

```python
gb2 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb2)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb2, annot = True)
plt.title(" Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

# from the above chart we have concluded that there is n0/negligible impact on the 
# student's score due to their parent's marital status

```python
gb3 = df.groupby("WklyStudyHours").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb3)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb3, annot = True)
plt.title(" Relationship between Weekly study hours and Student's Score")
plt.show()
```

# from the above chart we have concluded that there is n0/negligible impact on the 
# student's score due to their weekly study hours

```python
gb4 = df.groupby("PracticeSport").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb4)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb4, annot = True)
plt.title(" Relationship between Practice Sports and Student's Score")
plt.show()
```

# from the above chart we have concluded that there is n0 significance impact on the

# DISTRIBUTION OF ETHNIC GROUPS

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]

plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'EthnicGroup', hue = 'EthnicGroup')

for i in ax.containers:
    ax.bar_label(i,)
    
plt.show()
```

# from the above plots we can conclude that Group C has the maximum student count whereas Group A has the lowest student count
